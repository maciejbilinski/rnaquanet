```python
from numpy import percentile
q25, q70=percentile(df.rmsd.values, 25),percentile(df.rmsd.values, 70)
iqr = q70 - q25
cut_off = iqr * 1.5
upper =q75 + cut_off
print('Percentiles: 70th=%.3f, IQR=%.3f, upper=%.3f' % ( q75, iqr,upper))
# identify outliers
# remove outliers
f_df=df[df.rmsd<=upper]
f_train_df=train_df[train_df.rmsd<=upper]
f_val_df=val_df[val_df.rmsd<=upper]
f_test_df=test_df[test_df.rmsd<=upper]
# outliers_removed = [x for x in data if x >= lower and x <= upper]
# print('Non-outlier observations: %d' % len(outliers_removed))
```

### elements_no plots
#### overall

```python
a=df['elements_no'].value_counts().sort_index()
b=f_df['elements_no'].value_counts().sort_index()
c=((a-b)/a*100).fillna(100)
print((pd.concat([a.rename('Prefiltered'),b.rename('Filtered'),(c).rename('loss').apply(lambda x: "{:10.4f}%".format(x))],axis=1)).fillna(0))
plt.title('Elements in dataset (pre and post outlayer filtering)')
plt.bar(a.index,list(a))
plt.bar(b.index,list(b))
pass
```

```python
a=train_df['elements_no'].value_counts().sort_index()
b=f_train_df['elements_no'].value_counts().sort_index()
c=((a-b)/a*100).fillna(100)
print((pd.concat([a.rename('Prefiltered'),b.rename('Filtered'),(c).rename('loss').apply(lambda x: "{:10.4f}%".format(x))],axis=1)).fillna(0))
plt.title('Elements in train (pre and post outlayer filtering)')
plt.bar(a.index,list(a))
plt.bar(b.index,list(b))
pass
```
